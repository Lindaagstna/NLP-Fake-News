```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('stopwords', quiet=True)

stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

df = df.dropna(subset=['text'])

df['clean_text'] = (
    df['text']
    .str.lower()
    .str.replace(r'[^a-zA-Z\s]', '', regex=True)
    .str.split()
    .apply(lambda words: [stemmer.stem(w) for w in words if w not in stop_words])
    .str.join(' ')
)

df.head()
```

3. feature extraction

```python
# --- TF-IDF ---
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df['clean_text'])
print(f"Shape of X_tfidf: {X_tfidf.shape}")
```

```python
from gensim.models import Word2Vec

tokenized = df['clean_text'].apply(lambda x: x.split())

model_w2v = Word2Vec(sentences=tokenized, vector_size=100, window=5, min_count=1)

def get_avg_vector(tokens, model):
    vectors = []
    for word in tokens:
        if word in model.wv:
            vectors.append(model.wv[word])
    return np.mean(vectors, axis=0) if vectors else np.zeros(100)

X_w2v = np.array([get_avg_vector(tokens, model_w2v) for tokens in tokenized])
print(f"Shape of X_w2v: {X_w2v.shape}")
```

```python
from sklearn.model_selection import train_test_split

# Get the target variable
y = df['label']

# Create a boolean mask for rows where y is NaN
nan_mask = y.isna()

if nan_mask.any():
    print(f"Found {nan_mask.sum()} NaN values in the target variable 'y'. Removing corresponding rows.")
    # Filter y to remove NaN values
    y_filtered = y[~nan_mask]
    # Filter X_tfidf to remove corresponding rows
    X_tfidf_filtered = X_tfidf[~nan_mask.values]
    # Filter X_w2v to remove corresponding rows
    X_w2v_filtered = X_w2v[~nan_mask.values]
else:
    # If no NaNs, use the original variables
    y_filtered = y
    X_tfidf_filtered = X_tfidf
    X_w2v_filtered = X_w2v

# Perform the train-test split using the filtered data
X_train_tfidf, X_test_tfidf, y_train, y_test = train_test_split(X_tfidf_filtered, y_filtered, test_size=0.2, random_state=42)
X_train_w2v, X_test_w2v, _, _ = train_test_split(X_w2v_filtered, y_filtered, test_size=0.2, random_state=42)
```

4. modeling

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
nb = MultinomialNB()
lr = LogisticRegression(max_iter=1000)
svm = SVC()

nb.fit(X_train_tfidf, y_train)
lr.fit(X_train_tfidf, y_train)
svm.fit(X_train_tfidf, y_train)
```

```python
lr_w2v = LogisticRegression(max_iter=1000)
svm_w2v = SVC()

lr_w2v.fit(X_train_w2v, y_train)
svm_w2v.fit(X_train_w2v, y_train)
```

5. comporison model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def evaluate(model, X_test, y_test, name):
    y_pred = model.predict(X_test)
    print(f"=== {name} ===")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred, average='weighted'))
    print("Recall:", recall_score(y_test, y_pred, average='weighted'))
    print("F1-score:", f1_score(y_test, y_pred, average='weighted'))
    print()
```

```python
evaluate(nb, X_test_tfidf, y_test, "Naive Bayes (TF-IDF)")
evaluate(lr, X_test_tfidf, y_test, "Logistic Regression (TF-IDF)")
evaluate(svm, X_test_tfidf, y_test, "SVM (TF-IDF)")
```

```python
evaluate(lr_w2v, X_test_w2v, y_test, "Logistic Regression (Word2Vec)")
evaluate(svm_w2v, X_test_w2v, y_test, "SVM (Word2Vec)")
```

```python
import gradio as gr
import pickle
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# download resource nltk
nltk.download('stopwords')

# load model
model = pickle.load(open("model.pkl", "rb"))
tfidf = pickle.load(open("tfidf.pkl", "rb"))

# preprocessing
stemmer = PorterStemmer()
stop_words = set(stopwords.words("english"))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    words = text.split()
    words = [stemmer.stem(w) for w in words if w not in stop_words]
    return " ".join(words)

# fungsi prediksi
def predict_news(text):
    cleaned = clean_text(text)
    vector = tfidf.transform([cleaned])
    pred = model.predict(vector)[0]

    # kalau mau confidence
    try:
        prob = model.predict_proba(vector)[0]
        confidence = max(prob)
    except:
        confidence = None

    label = "🟩 REAL" if pred == 1 else "🟥 HOAX"

    if confidence:
        return f"{label} ({confidence:.2f})"
    return label

# UI Gradio (PAKE GRADIENT)
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# 📰 Fake News Detection")
    gr.Markdown("Masukkan teks berita untuk mengecek apakah hoax atau real")

    with gr.Row():
        text_input = gr.Textbox(lines=5, placeholder="Masukkan berita di sini...")

    btn = gr.Button("Cek Sekarang 🚀")

    output = gr.Textbox(label="Hasil")

    btn.click(fn=predict_news, inputs=text_input, outputs=output)

# launch
demo.launch()
```
